Dropping Irrelavant Columns

In new york, time of date (seasonal) can effect sales, so month and dates can be valuable. While hour may seem irrelavant - it can be used to determine which hours the business receives the most traffic (maybe during after school hours or late night purchases). Address is far less important than the generalized city and the Order ID doesn't seem necessary. The price of each is also less important than the sale value, and since Price Each and Quantity Ordered are linked by contributing to sale, we can drop them both.

```python
df = df.drop(["Purchase Address", "Price Each", "Quantity Ordered", "Order ID"], axis =1 )
```

```python
df.City.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Sales of Cities")
plt.ylabel('Sales')
plt.xlabel('City');
```
